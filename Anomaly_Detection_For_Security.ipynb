```python
%pylab inline
import pandas as pd
from sklearn.svm import SVDD
from sklearn.datasets import load_svmlight_file
from sklearn.cross_validation import train_test_split
from sklearn.metrics import roc_curve, auc, accuracy_score, precision_score
from pandas.tools.plotting import scatter_matrix
```

```python
X_train, y_train = load_svmlight_file("Training Data-Intrusion Detection Dataset.txt")
column_names = ['NUM SAME SA DA DP', 'RATE DIFF ALERT SAME SA DA DP', 'TIME STDDEV SAME SA DA DP', 'NUM SAME SA DP DIFF DA', 
                'RATE DIFF ALERT SAME SA DP DIFF DA', 'TIME STDDEV SAME SA DP DIFF DA', 'RATE REVERSE SP SAME SA DP']

data = pd.DataFrame(X_train.todense(), columns=column_names)
data['label'] = y_train

#data.drop_duplicates(inplace=True)
```

### Duplicates Data

```python
non_duplicates = data.drop_duplicates()
print "Rows count = {}".format(len(data))
print "Rows count after deleting duplicates = {}".format(len(non_duplicates.index))
duplicates = data.index  - non_duplicates.index
print "Duplicate rows count = {}".format(len(duplicates))
duplicates = data.iloc[duplicates,:]
```

Target label distribution in duplicate data

```python
scatter_matrix(duplicates, alpha=0.2, figsize=(20, 20), diagonal='hist')
```

Target label distribution in non dulicate data

```python
scatter_matrix(data.drop_duplicates(), alpha=0.2, figsize=(20, 20), diagonal='hist')
```

# With duplicate rows

Separate train and test data

```python
train_x, test_x, train_y, test_y = train_test_split(data.drop('label', axis=1), data['label'], test_size=0.5, random_state=0)
```

#### RBF kernel

```python
clf = SVDD(C = 1, kernel='rbf', gamma=0.01)
```

Select clean data

```python
clean_data = train_x[(train_y == 1).nonzero()[0]]
```

```python
clf.fit(clean_data)
```

```python
prediction = clf.predict(test_x)
```

```python
print "False Alarm = {}".format(mean((prediction == -1) * (test_y == 1)))
```

```python
print "Miss Target = {}".format(mean((prediction == 1) * (test_y == -1)))
```

```python
print "Precision = {}".format(precision_score(test_y, prediction))
```

```python
print "Acuuracy Score = {}".format(accuracy_score(test_y, prediction))
```

```python
prediction = clf.decision_function(test_x)
```

```python
fpr, tpr, _ = roc_curve(test_y, prediction)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
legend(loc='best')
```
